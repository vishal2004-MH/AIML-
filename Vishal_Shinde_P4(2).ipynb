```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
house_data = pd.read_csv("house_data.csv")
```

```python
house_data.head()
```

```python
#Extract relevant features and target variable
X = house_data[['Square_Footage', 'Bedrooms']] # Features
y = house_data['Price'] # Target variables
```

```python
#Training Data Preparation:
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20)
```

```python
# Display the training and testing data
print("Training Data:")
print(pd.concat([X_train, y_train], axis=1)) # Concatenate features
```

```python
print("\nTesting Data:")
print(pd.concat([X_test, y_test], axis=1)) # Concatenate features
```

```python
# Create figure
plt.figure(figsize=(10, 5))

# Plot training data
plt.scatter(X_train['Square_Footage'], X_train['Bedrooms'], color='blue', label='Training Data')

# Set plot labels and title
plt.xlabel('Square Footage')
plt.ylabel('Bedrooms')
plt.title('Training and Testing Data')

# Add legend
plt.legend()

# Show plot
plt.show()
```

```python
# Plot testing data only
plt.scatter(X_test['Square_Footage'], X_test['Bedrooms'], color='red', label='Testing Data')

# Set labels and title
plt.xlabel('Square Footage')
plt.ylabel('Bedrooms')
plt.title('Testing Data')

# Add legend
plt.legend()

# Show plot
plt.show()
```
