**Pro 3 -  To extract features from given dataset and establish training data.**

```python
# Import the relevant libraries
import numpy as np
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
```

```python
# # Generating a random dataset with 2 features and 100 samples
np.random.seed(0)
X = np.random.rand(20, 100)
print("Original dataset:\n", X)

```

```python
# Applying PCA
pca = PCA(n_components=20)
principal_components = pca.fit_transform(X)
```

```python
# Printing the principal components
print("Principal Components:\n", principal_components)
```

```python
# Printing the principal component vectors (eigenvectors)
print("Principal Component Vectors (Eigenvectors):\n", pca.components_)
```

```python
# Printing the mean of the data
print("Mean of the Data:\n", pca.mean_)
```

```python
# Generating the PCA equations
for i in range(pca.components_.shape[0]):
 equation = f"PC{i+1} = "
 equation += " + ".join([f"({pca.components_[i, j]:.4f} * X{j+1})" for j in range(pca.components_.shape[1])])
 equation += f" + ({pca.mean_[i]:.4f})"
print(equation)
```

```python
# Plotting the original data
plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], color='blue', label='Original Data')
# Plotting the principal components
plt.scatter(principal_components[:, 0], principal_components[:, 1], color='green', label='Principal Components')
# Calculating and plotting the principal component vectors
origin = np.mean(X[:, :2], axis=0) # Calculate the mean for the first two features only
for length, vector, label in zip(pca.explained_variance_, pca.components_, ['PC1', 'PC2']):
  v = vector * 2 * np.sqrt(length) # Scale the vectors for better visualization
  plt.quiver(origin[0], origin[1], v[0], v[1], angles='xy', scale_units='xy', scale=1, color='red', width=0.005)  # Pass origin as two separate values
  plt.text(origin[0] + v[0], origin[1] + v[1], label, color='red', fontsize=12, ha='center')
# Plot settings
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('Original Data, Principal Components, and Principal Component Vectors')
plt.legend()
plt.grid()
plt.show()
```

```python
# Explained variance ratio
print("Explained Variance Ratio:\n", pca.explained_variance_ratio_)
```

The explained variance ratio tells us how much of the total variance in the data is captured by each principal component. In this case, the sumation of explained variance ratios for the first sixteen principal components is 0.923.

Conclusion:
The first sixteen principal components (PC1 to PC16) captures about 92.3% of the total variance in the data .
Since first sixteen components are needed to adequately represent the structure of data.
The dimensionality of Data set is reduced from 100 to 20.


```python
# To establish training data
# Importing the relevant library
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
# Considering most important Principal components
X_pca = principal_components [: , 0:16]
```

```python
# Creating synthetic target labels for demonstration purposes
# For example, let's create binary labels based on some condition
y = np.where(X[:, 0] + X[:, 1] > 1, 1, 0)
y
```

```python
# Splitting the PCA-transformed data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_pca, y, test_size=0.2, random_state=0)
# Printing the shapes of the training and testing sets
print("Training Set Shape (X_train):", X_train.shape)
print("Training Set Shape (y_train):", y_train.shape)
print("Testing Set Shape (X_test):", X_test.shape)
print("Testing Set Shape (y_test):", y_test.shape)
```

```python
# Printing the training data
print('The training data for X is \n\n', X_train)
print('\n\nThe training data for y is \n\n', y_train)
```
